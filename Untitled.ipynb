```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot(x, y, title=""):
    plt.title(title)
    plt.axes().set_aspect("equal", "datalim")
    plt.grid()
    plt.plot(x, y, "black")
    plt.show()
```

## 2018年度作品

### 首都大学東京

```python
x = np.arange(-5, 5, 0.001)
def syutodai(x):
    y_1 = (1.2 * np.exp(-0.5 * x ** 2))
    y_2 = (np.exp(0.1 * x) - 1.3 ** (0.6 * x) - 1.1 ** -x)
    y_3 = (0.2 * np.exp(-7 * (x - 0.8) ** 8))
    y_4 = (0.1 * np.exp(-(8 * x - 2) ** 2) - 1.1 ** (x - 6))
    y_5 = (0.4 * np.exp(-0.04 * x ** 8))
    y_6 = (0.4 * np.exp(-0.04 * (x + 3) ** 8))
    y_7 = (1.2)
    return y_1 + y_2 + y_3 + y_4 + y_5 + y_6 + y_7
y = syutodai(x)
plot(x, y, title="syutodai")
```

```python
plot(y[::-1], x, title="syutodai")
```

### 大阪大学

```python
x = np.arange(-2, 2, 0.001)
def handai(x):
    e = np.e
    y_1 = ((np.sqrt((np.absolute(1 - x) + 1 - x) / 2) + (1 / 4)) * np.exp(-(1 - x) ** 2))
    y_2_1 = (1 / 40)
    y_2_2 = (np.absolute(1 - 2 ** 4 * (5 * x - 3) ** 4))
    y_2_3 = (np.absolute(1 - 2 ** 9 * (5 * x - 3) ** 4))
    y_2_4 = (2 - 528 * (5 * x - 3) ** 4)
    y_2 = (y_2_1 * (y_2_2 + y_2_3 + y_2_4))
    y = y_1 + y_2

    return y
y = handai(x)

plot(x, y, title="handai")
```

```python
plot(y[::-1], x, title="handai")
```

### 名古屋大学

```python
x = np.arange(-3, 3, 0.001)
def nagoya(x):
    y_1 = (1.5 * np.exp(-0.62 * (x - 0.16) ** 2)) / (1 + np.exp(-20 * (5 * x - 1)))
    y_2 = (1.5 + 0.8 * (x - 0.2) ** 3) * (1 + np.exp(20 * (5 * x - 1))) ** -1
    y_3 = (1 + np.exp(-(100 * (x + 1) - 16)))
    y_4 = (0.2 * (np.exp(-(x + 1) ** 2) + 1)) / (1 + np.exp(100 * (x + 1) - 16))
    y_5 = (0.1 / np.exp(2 * (10 * x - 1.2) ** 4))
    y = y_1 + (y_2 / y_3) + y_4 + y_5

    return y
y = nagoya(x)
plot(x, y, title="nagoya")
```

```python
plot(x[::-1], y, title="nagoya")
```
